```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use('seaborn-darkgrid')

file_path = 'Financial_Report_2004-2023.xlsx'

intel_df = pd.read_excel(file_path, sheet_name='Financial_Report_Intel')
amd_df = pd.read_excel(file_path, sheet_name='Financial_Report_AMD')
tsmc_df = pd.read_excel(file_path, sheet_name='Financial_Report_TSMC')
nvidia_df = pd.read_excel(file_path, sheet_name='Financial_Report_NVIDIA')

fig, axs = plt.subplots(2, 1, figsize=(14, 12), sharex=True)
axs[0].plot(intel_df['Year'], intel_df['Net_revenue'], label='Intel Revenue', marker='o', color='blue')
axs[0].plot(amd_df['Year'], amd_df['Net_revenue'], label='AMD Revenue', marker='o', color='purple')
axs[0].plot(tsmc_df['Year'], tsmc_df['Net_revenue'], label='TSMC Revenue', marker='o', color='green')
axs[0].plot(nvidia_df['Year'], nvidia_df['Net_revenue'], label='NVIDIA Revenue', marker='o', color='red')
axs[0].set_title('Revenue (2004-2023)')
axs[0].set_ylabel('Revenue ($ Millions)')
axs[0].legend()
axs[0].grid(True)
axs[1].plot(intel_df['Year'], intel_df['Net_income'], label='Intel Net Income', marker='o', color='blue')
axs[1].plot(amd_df['Year'], amd_df['Net_income'], label='AMD Net Income', marker='o', color='purple')
axs[1].plot(tsmc_df['Year'], tsmc_df['Net_income'], label='TSMC Net Income', marker='o', color='green')
axs[1].plot(nvidia_df['Year'], nvidia_df['Net_income'], label='NVIDIA Net Income', marker='o', color='red')
axs[1].set_title('Net Income (2004-2023)')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Net Income ($ Millions)')
axs[1].legend()
axs[1].grid(True)
plt.xticks(intel_df['Year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculating R&D and Marketing, General, and Administrative expenses as a percentage of revenue for each company
intel_df['R&D_Percentage'] = (intel_df['Research_and_development'] / intel_df['Net_revenue']) * 100
intel_df['Marketing_General_Admin_Percentage'] = (intel_df['Marketing_general_and_administrative'] / intel_df['Net_revenue']) * 100

amd_df['R&D_Percentage'] = (amd_df['Research_and_development'] / amd_df['Net_revenue']) * 100
amd_df['Marketing_General_Admin_Percentage'] = (amd_df['Marketing_general_and_administrative'] / amd_df['Net_revenue']) * 100

tsmc_df['R&D_Percentage'] = (tsmc_df['Research_and_development'] / tsmc_df['Net_revenue']) * 100
tsmc_df['Marketing_General_Admin_Percentage'] = (tsmc_df['Marketing_general_and_administrative'] / tsmc_df['Net_revenue']) * 100

nvidia_df['R&D_Percentage'] = (nvidia_df['Research_and_development'] / nvidia_df['Net_revenue']) * 100
nvidia_df['Marketing_General_Admin_Percentage'] = (nvidia_df['Marketing_general_and_administrative'] / nvidia_df['Net_revenue']) * 100

fig, axs = plt.subplots(2, 1, figsize=(14, 12), sharex=True)
axs[0].plot(intel_df['Year'], intel_df['R&D_Percentage'], label='Intel R&D % of Revenue', marker='o', color='blue')
axs[0].plot(amd_df['Year'], amd_df['R&D_Percentage'], label='AMD R&D % of Revenue', marker='o', color='purple')
axs[0].plot(tsmc_df['Year'], tsmc_df['R&D_Percentage'], label='TSMC R&D % of Revenue', marker='o', color='green')
axs[0].plot(nvidia_df['Year'], nvidia_df['R&D_Percentage'], label='NVIDIA R&D % of Revenue', marker='o', color='red')
axs[0].set_title('R&D as Percentage of Revenue (2004-2023)')
axs[0].set_ylabel('R&D % of Revenue')
axs[0].legend()
axs[0].grid(True)
axs[1].plot(intel_df['Year'], intel_df['Marketing_General_Admin_Percentage'], label='Intel MG&A % of Revenue', marker='o', color='blue')
axs[1].plot(amd_df['Year'], amd_df['Marketing_General_Admin_Percentage'], label='AMD MG&A % of Revenue', marker='o', color='purple')
axs[1].plot(tsmc_df['Year'], tsmc_df['Marketing_General_Admin_Percentage'], label='TSMC MG&A % of Revenue', marker='o', color='green')
axs[1].plot(nvidia_df['Year'], nvidia_df['Marketing_General_Admin_Percentage'], label='NVIDIA MG&A % of Revenue', marker='o', color='red')
axs[1].set_title('MG&A as Percentage of Revenue (2004-2023)')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('MG&A % of Revenue')
axs[1].legend()
axs[1].grid(True)

plt.xticks(intel_df['Year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(4, 1, figsize=(10, 20))

# Plot titles
titles = ['Intel', 'AMD', 'TSMC', 'NVIDIA']

for i, (df, title) in enumerate(zip([intel_df, amd_df, tsmc_df, nvidia_df], titles)):
    axs[i].plot(df['Year'], df['R&D_Percentage'], label='R&D % of Revenue', marker='o', color='blue')
    axs[i].plot(df['Year'], df['Marketing_General_Admin_Percentage'], label='MG&A % of Revenue', marker='o', color='red')
    axs[i].set_title(f'{title}')
    axs[i].set_ylabel('% of Revenue')
    axs[i].legend()
    axs[i].grid(True)
    axs[i].set_xticks(df['Year'])
    axs[i].set_xticklabels(df['Year'], rotation=45)

axs[-1].set_xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
# Calculating correlation coefficients between R&D spending and Net Income for each company
rd_net_income_corr = {
    "Intel": intel_df['Research_and_development'].corr(intel_df['Net_income']),
    "AMD": amd_df['Research_and_development'].corr(amd_df['Net_income']),
    "TSMC": tsmc_df['Research_and_development'].corr(tsmc_df['Net_income']),
    "NVIDIA": nvidia_df['Research_and_development'].corr(nvidia_df['Net_income'])
}

# Calculating correlation coefficients between MG&A expenses and Net Income for each company
mga_net_income_corr = {
    "Intel": intel_df['Marketing_general_and_administrative'].corr(intel_df['Net_income']),
    "AMD": amd_df['Marketing_general_and_administrative'].corr(amd_df['Net_income']),
    "TSMC": tsmc_df['Marketing_general_and_administrative'].corr(tsmc_df['Net_income']),
    "NVIDIA": nvidia_df['Marketing_general_and_administrative'].corr(nvidia_df['Net_income'])
}

correlation_data = pd.DataFrame({
    'Company': ['Intel', 'AMD', 'TSMC', 'NVIDIA'] * 2,
    'Correlation Coefficient': list(rd_net_income_corr.values()) + list(mga_net_income_corr.values()),
    'Type': ['R&D-Net Income'] * 4 + ['MG&A-Net Income'] * 4
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Company', y='Correlation Coefficient', hue='Type', data=correlation_data, palette=["#800080", "#0000FF"])
plt.title('Correlation between R&D/MG&A Spending and Net Income')
plt.ylabel('Correlation Coefficient')
plt.xlabel('Company')
plt.legend()
plt.tight_layout()
plt.show()

correlation_data.pivot(index='Company', columns='Type', values='Correlation Coefficient')
```

```python
# Define the years for the crisis and pandemic periods
crisis_years = [2008, 2009, 2010, 2011]
pandemic_years = [2020, 2021, 2022, 2023]

# Filtering data for crisis and pandemic periods
crisis_data = [df[df['Year'].isin(crisis_years)] for df in [intel_df, amd_df, tsmc_df, nvidia_df]]
pandemic_data = [df[df['Year'].isin(pandemic_years)] for df in [intel_df, amd_df, tsmc_df, nvidia_df]]
companies = ['Intel', 'AMD', 'TSMC', 'NVIDIA']

# Calculating correlation coefficients
crisis_corrs_rd = [df['Research_and_development'].corr(df['Net_income']) for df in crisis_data]
crisis_corrs_mga = [df['Marketing_general_and_administrative'].corr(df['Net_income']) for df in crisis_data]
pandemic_corrs_rd = [df['Research_and_development'].corr(df['Net_income']) for df in pandemic_data]
pandemic_corrs_mga = [df['Marketing_general_and_administrative'].corr(df['Net_income']) for df in pandemic_data]

correlation_df = pd.DataFrame({
    'Company': companies * 4,
    'Correlation': crisis_corrs_rd + crisis_corrs_mga + pandemic_corrs_rd + pandemic_corrs_mga,
    'Period': ['Crisis R&D'] * 4 + ['Crisis MG&A'] * 4 + ['Pandemic R&D'] * 4 + ['Pandemic MG&A'] * 4
})

companies = ['Intel', 'AMD', 'TSMC', 'NVIDIA']

fig, ax = plt.subplots(figsize=(12, 7))
sns.barplot(x='Company', y='Correlation', hue='Period', data=correlation_df, palette=['#6A1B9A', '#9C27B0', '#1565C0', '#42A5F5'], edgecolor='white')
ax.set_title('Expense Correlation with Net Income During 2008 Crisis and COVID-19 Pandemic', fontsize=16, fontweight='bold')
ax.set_ylabel('Correlation Coefficient', fontsize=14, fontweight='bold')
ax.set_xlabel('Company')
ax.set_xticklabels(companies, fontsize=12, fontweight='bold')
ax.legend(frameon=True, loc='lower right', fontsize=12)
ax.set_xticklabels(companies, fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
fig.savefig('Correlation_of_R&D_and_MG&A_Expenses_with_Net_Income.png', dpi=300, format='png')
```

```python
# Function to calculate correlations for given years within a company's dataframe
def calculate_correlations(df, years):
    filtered_df = df[df['Year'].isin(years)]
    r_and_d_corr = filtered_df['Research_and_development'].corr(filtered_df['Net_income'])
    mg_and_a_corr = filtered_df['Marketing_general_and_administrative'].corr(filtered_df['Net_income'])
    return r_and_d_corr, mg_and_a_corr

# Calculate correlations for crisis and pandemic periods for each company
intel_crisis_corr = calculate_correlations(intel_df, crisis_years)
amd_crisis_corr = calculate_correlations(amd_df, crisis_years)
tsmc_crisis_corr = calculate_correlations(tsmc_df, crisis_years)
nvidia_crisis_corr = calculate_correlations(nvidia_df, crisis_years)

intel_pandemic_corr = calculate_correlations(intel_df, pandemic_years)
amd_pandemic_corr = calculate_correlations(amd_df, pandemic_years)
tsmc_pandemic_corr = calculate_correlations(tsmc_df, pandemic_years)
nvidia_pandemic_corr = calculate_correlations(nvidia_df, pandemic_years)

amd_crisis_corr, amd_pandemic_corr
```
